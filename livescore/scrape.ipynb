```python
import urllib.request
import json

url = f'https://livescore-api.com/api-client/matches/history.json?key={API_KEY}&secret={API_SECRET}&competition_id=4'
req = urllib.request.Request(url)
response = urllib.request.urlopen(req)

# Read and decode the response
data = response.read().decode('utf-8')

# Parse JSON and display with indentation
json_data = json.loads(data)
print(json.dumps(json_data, indent=2))
```

```python
def make_request(endpoint, params=None):
    if params is None:
        params = {}
    params.update({'key': API_KEY, 'secret': API_SECRET})
    url = f"{BASE_URL}{endpoint}"
    response = requests.get(url, params=params)
    response.raise_for_status()
    return response.json()
```

```python
# === STEP 1: Get countries ===
def get_countries_json():
    """
    Returns JSON with list of all supported countries
    """
    return make_request("/countries/list.json")

# Call the function and display the JSON
try:
    countries_data = get_countries_json()
    print(json.dumps(countries_data, indent=2))
except Exception as e:
    print(f"Error fetching countries: {e}")
```

```python
# === STEP 2: Get competitions (leagues & cups) for a country ===
def get_competitions_json(country_id):
    """
    Returns JSON of competitions for the given country_id
    """
    return make_request("/competitions/list.json", {'country_id': country_id})

# Call the function with England's ID (19) and display the JSON
try:
    england_competitions = get_competitions_json("19")
    print(json.dumps(england_competitions, indent=2))
except Exception as e:
    print(f"Error fetching competitions for England: {e}")
```

```python
# === STEP 3: Get seasons for a competition ===
def get_seasons_json(comp_id):
    """
    Returns JSON of seasons for the given competition ID
    """
    return make_request("/seasons/list.json", {'comp_id': comp_id})

# Call the function with Premier League's ID (2) and display the JSON
try:
    premier_league_seasons = get_seasons_json("2")
    print(json.dumps(premier_league_seasons, indent=2))
except Exception as e:
    print(f"Error fetching seasons for Premier League: {e}")
```

```python
# === STEP 4: Get historical match data ===
def get_history_matches_json(comp_id, from_date=None, to_date=None, page=1):
    """
    Returns JSON of finished matches for a given competition using the correct /matches/history.json endpoint.
    You can pass from_date and to_date optionally in YYYY-MM-DD format.
    """
    params = {
        'competition_id': comp_id,
        'page': page
    }
    if from_date:
        params['from'] = from_date
    if to_date:
        params['to'] = to_date

    return make_request("/matches/history.json", params)


# Call the function with Premier League ID (2) for a date range
try:
    premier_league_matches = get_history_matches_json(
        comp_id="2",
        from_date="2024-08-01",
        to_date="2025-07-27",
        page=1
    )
    print(json.dumps(premier_league_matches, indent=2))
except Exception as e:
    print(f"Error fetching Premier League matches: {e}")
```

```python
# === STEP 5: Get match statistics ===
def get_match_statistics_json(match_id):
    """
    Returns JSON match statistics for a given match ID
    """
    return make_request("/matches/stats.json", {'match_id': match_id})

# Call the function with match ID (627919) and display the JSON
try:
    match_stats = get_match_statistics_json("627919")
    print(json.dumps(match_stats, indent=2))
except Exception as e:
    print(f"Error fetching match statistics for match 627919: {e}")
```

```python
# === Fixed: Complete pipeline example ===
def run_full_example():
    try:
        # Step 1: Get England competitions directly
        print(" Step 1: Getting England competitions...")
        comps_json = get_competitions_json("19")  # England ID
        competitions = comps_json['data']['competition']
        print(f" Found {len(competitions)} competitions for England")
        
        # Find Premier League (ID: 2)
        premier_league = next((c for c in competitions if c['id'] == 2), None)
        if not premier_league:
            print(" Premier League not found")
            return None
        
        print(f" Found Premier League: {premier_league['name']}")
        
        # Step 2: Get Premier League seasons
        print("\n Step 2: Getting Premier League seasons...")
        seasons_json = get_seasons_json("2")  # Premier League ID
        seasons = seasons_json['data']['season']
        print(f" Found {len(seasons)} seasons for Premier League")
        
        # Step 3: Get recent matches using direct API call (like your working version)
        print("\n Step 3: Getting recent Premier League matches...")
        try:
            # Use the working approach from your cell 8
            history_json = make_request("/matches/history.json", {
                'competition_id': '2',
                'from': '2024-08-01',
                'to': '2025-07-27',
                'page': 1
            })
            matches = history_json['data']['match']
            print(f" Found {len(matches)} matches")
        except Exception as e:
            print(f" Failed to get matches: {e}")
            return None
        
        # Step 4: Get statistics for the first match
        if matches:
            first_match = matches[0]
            match_id = first_match['id']
            print(f"\n Step 4: Getting statistics for {first_match['home_name']} vs {first_match['away_name']}...")
            
            stats_json = get_match_statistics_json(str(match_id))
            print(f" Stats fetched for match ID {match_id}")
            
            return {
                'competitions': comps_json,
                'seasons': seasons_json,
                'matches': history_json,
                'sample_stats': stats_json
            }
        else:
            print(" No matches found")
            return None
            
    except Exception as e:
        print(f" Error in pipeline: {e}")
        import traceback
        traceback.print_exc()
        return None

# Run the fixed pipeline
print(" Starting Livescore API Pipeline...")
result = run_full_example()

if result:
    print("\n Pipeline completed successfully!")
    print(" Data collected:")
    for key, value in result.items():
        print(f"  - {key}: ")
else:
    print("\n Pipeline failed!")
```
