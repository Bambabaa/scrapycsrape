```python
def make_request(endpoint, params=None):
    if params is None:
        params = {}
    params.update({'key': API_KEY, 'secret': API_SECRET})
    url = f"{BASE_URL}{endpoint}"
    response = requests.get(url, params=params)
    response.raise_for_status()
    return response.json()
```

```python
# === STEP 1: Get countries ===
def get_countries_json():
    """
    Returns JSON with list of all supported countries
    """
    return make_request("/countries/list.json")

# Call the function and display the JSON
try:
    countries_data = get_countries_json()
    print(json.dumps(countries_data, indent=2))
except Exception as e:
    print(f"Error fetching countries: {e}")
```

```python
# === STEP 2: Get competitions (leagues & cups) for a country ===
def get_competitions_json(country_id):
    """
    Returns JSON of competitions for the given country_id
    """
    return make_request("/competitions/list.json", {'country_id': country_id})

# Call the function with England's ID (19) and display the JSON
try:
    england_competitions = get_competitions_json("19")
    print(json.dumps(england_competitions, indent=2))
except Exception as e:
    print(f"Error fetching competitions for England: {e}")
```

```python
# === STEP 3: Get seasons for a competition ===
def get_seasons_json(comp_id):
    """
    Returns JSON of seasons for the given competition ID
    """
    return make_request("/seasons/list.json", {'comp_id': comp_id})

# Call the function with Premier League's ID (2) and display the JSON
try:
    premier_league_seasons = get_seasons_json("2")
    print(json.dumps(premier_league_seasons, indent=2))
except Exception as e:
    print(f"Error fetching seasons for Premier League: {e}")
```

```python
# === STEP 4: Get historical match data ===
def get_history_matches_json(comp_id, from_date=None, to_date=None, page=1):
    """
    Returns JSON of finished matches for a given competition using the correct /matches/history.json endpoint.
    You can pass from_date and to_date optionally in YYYY-MM-DD format.
    """
    params = {
        'competition_id': comp_id,
        'page': page
    }
    if from_date:
        params['from'] = from_date
    if to_date:
        params['to'] = to_date

    return make_request("/matches/history.json", params)


# Call the function with Premier League ID (2) for a date range
try:
    premier_league_matches = get_history_matches_json(
        comp_id="2",
        from_date="2024-08-01",
        to_date="2025-07-27",
        page=1
    )
    print(json.dumps(premier_league_matches, indent=2))
except Exception as e:
    print(f"Error fetching Premier League matches: {e}")
```

```python
# === STEP 5: Get match statistics ===
def get_match_statistics_json(match_id):
    """
    Returns JSON match statistics for a given match ID
    """
    return make_request("/matches/stats.json", {'match_id': match_id})

# Call the function with match ID (627919) and display the JSON
try:
    match_stats = get_match_statistics_json("627919")
    print(json.dumps(match_stats, indent=2))
except Exception as e:
    print(f"Error fetching match statistics for match 627919: {e}")
```

```python
def get_combined_match_data(comp_id, from_date=None, to_date=None, page=1, max_matches=5):
    all_data = []

    matches_json = get_history_matches_json(comp_id, from_date, to_date, page)
    matches = matches_json.get('data', {}).get('match', [])

    if not isinstance(matches, list) or len(matches) == 0:
        print(" No valid matches returned. Here's what we got:")
        print(json.dumps(matches_json, indent=2))
        return []

    for match in matches[:max_matches]:
        match_id = match['id']
        try:
            stats_json = get_match_statistics_json(match_id)
            stats = stats_json.get("data", {})
        except Exception as e:
            print(f"Error fetching stats for match {match_id}: {e}")
            stats = {}

        # Use your preferred format with parse_stat function
        def parse_stat(stat_str):
            if stat_str and ':' in stat_str:
                h, a = stat_str.split(':')
                return {"home": int(h), "away": int(a)}
            return {"home": None, "away": None}
        
        
        combined = {
            "match_id": match["id"],
            "fixture_id": match.get("fixture_id"),
            "date": match["date"],
            "status": match["status"],
            "competition": match["competition"]["name"],
            "round": match.get("round"),
            "venue": match.get("location"),
            "home_team": match["home"]["name"],
            "away_team": match["away"]["name"],
            "score": match["scores"]["score"],
            "ht_score": match["scores"]["ht_score"],
            "ft_score": match["scores"]["ft_score"],
            "outcome": match.get("outcomes", {}).get("full_time"),

            "stats": {
                "possession": parse_stat(stats.get("possesion")),
                "shots_on_target": parse_stat(stats.get("shots_on_target")),
                "shots_off_target": parse_stat(stats.get("shots_off_target")),
                "attempts_on_goal": parse_stat(stats.get("attempts_on_goal")),
                "shots_blocked": parse_stat(stats.get("shots_blocked")),
                "corners": parse_stat(stats.get("corners")),
                "offsides": parse_stat(stats.get("offsides")),
                "yellow_cards": parse_stat(stats.get("yellow_cards")),
                "red_cards": parse_stat(stats.get("red_cards")),
                "fouls": parse_stat(stats.get("fauls")),
                "saves": parse_stat(stats.get("saves")),
                "dangerous_attacks": parse_stat(stats.get("dangerous_attacks")),
                "attacks": parse_stat(stats.get("attacks")),
                "penalties": parse_stat(stats.get("penalties")),

                "substitutions": parse_stat(stats.get("substitutions")),
                "free_kicks": parse_stat(stats.get("free_kicks")),
                "goal_kicks": parse_stat(stats.get("goal_kicks")),
                "throw_ins": parse_stat(stats.get("throw_ins")),
                "treatments": parse_stat(stats.get("treatments"))
                
            },

            "odds": match.get("odds", {}).get("pre", {})
        }

        all_data.append(combined)

    return all_data


if __name__ == "__main__":
    try:
        data = get_combined_match_data(
            comp_id="2", 
            from_date="2024-08-01", 
            to_date="2025-01-27", 
            page=1, 
            max_matches=3
        )
        print(json.dumps(data, indent=2))
    except Exception as e:
        print(f"Error: {e}")
```
